```python
corr_matrix = bos.corr()
sns.set(rc={'figure.figsize':(11.7,8.27)})

sns.heatmap(corr_matrix, annot=True, )
```

```python
X = bos.drop(columns=['MEDV'])
y = bos['MEDV']
plt.figure(figsize=(20,25),facecolor="white")
plotnumber = 1

for column in X:
    if plotnumber <= 16:
        ax = plt.subplot(5,3,plotnumber)
        sns.scatterplot(X[column],y)
        plt.xlabel(column)
    plotnumber+=1

plt.tight_layout()
```

### Feature selection
From above observation, two features RM and LSAT are very much linear with repect to target column MEDV.
From the heat map we can see that RM is positively correlated and LSAT is negatively correlated with 0.7 and -0.74 respectively.

```python
x = pd.DataFrame(pd.concat([bos['RM'],bos['LSTAT']], axis=1), columns=['RM', 'LSTAT'])
x.head()
```

```python
# Spiltting the data 
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
print(x_train.head())
print(x_test.head())
print(y_train.head())
print(y_test.head())
```

```python
# Training the model with linear regression algo

from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV, ElasticNet, ElasticNetCV

lr =  LinearRegression()
mymodel = lr.fit(x_train, y_train)
```

```python
# R2 of train dataset
mymodel.score(x_train,y_train)
```

```python
def adjustedR2(x,y):
    R2 = mymodel.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjustedR2 = 1 - (1-R2)*(n-1)/(n-p-1)
    return adjustedR2
```

```python
# adjusted R2 of train dataset
adjustedR2(x_train, y_train)
```

```python
# Evaluating if overfitting with test sets
# R2 of test dataset
mymodel.score(x_test,y_test)
```

```python
# adjusted R2 of test dataset
adjustedR2(x_test,y_test)
```

```python
# Lasso Regularization
# LassoCV will return best alpha and coefficients after performing 10 cross validations
lasscv = LassoCV(alphas = None,cv =10, max_iter = 100000, normalize = True)
lasscv.fit(x_train, y_train)
```

```python
alpha=lasscv.alpha_
alpha
```

```python
lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

### Score for test data on Lasso regularization model and mymodel are same thus our model is not overfitted.

```python
#prediction with greater LSTAT
mymodel.predict([[2.6,5]])
```

```python
#prediction with lower LSTAT
mymodel.predict([[2.6,1]])
```

So the predicted prize is decreasing when LSTAT is higher which was observed in the above graph.
